```python
# imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# load dataset
df = pd.read_csv(r'dataset/transformer-voltage.csv',
                index_col='Sample',
                delimiter=",")
df
```

```python
# summary of dataframe
df.info()
```

```python
# summary of statistics
df.describe()
```

```python
# z-critical for two-tailed test
alpha = 0.05
z_critical = stats.norm.ppf(1-alpha/2)
z_critical
```

```python
# confidence interval
pop_std = 5

ci_lower, ci_upper = stats.norm.interval(
    confidence = 0.95,
    loc = df['Voltage'].mean(),
    scale = pop_std/np.sqrt(df['Voltage'].count())
)

print(f"CI: [{ci_lower:.2f},{ci_upper:.2f}]")
```

```python
# voltage distribution
mean = df ['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

# x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

# normal distribution plot
x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

# histogram
plt.hist(df['Voltage'], bins=5,
            density = True,
            alpha = 0.8,
            color='green',
            edgecolor='Black')

plt.plot(x, p,
        color='blue',
        linewidth=2)

plt.title('Voltage Distribution',
            font='Tahoma',
            fontsize=12)
plt.xlabel('Voltage (V)')
plt.ylabel('Probability Density')

# confidence interval
plt.axvline(x=ci_lower,
            color='red',
            linestyle='--')
plt.axvline(x=ci_upper,
            color='red',
            linestyle='--')
plt.show()
```
